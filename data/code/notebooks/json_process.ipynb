```python
def process_all_docs(folder,exclude_type,img_store,max_pages=50):
    total_docs = []
    folders = os.listdir(folder)
    folders_filt = [i for i in folders if i not in exclude_type and os.path.isdir(os.path.join(folder,i))]
    n = 0
    for i in folders_filt:
        print(f'processing {i}... {(n+1)}/{len(folders_filt)}')
        path_i = os.path.join(folder,i)
        folders_i = os.listdir(path_i)
        folders_i_filt = [t for t in folders_i if os.path.isdir(os.path.join(folder,i,t))]
        k = 0
        for j in folders_i_filt:
            print(f'processing sub_folder {j} - {k+1}/{len(folders_i_filt)}\n')
            j_path = os.path.join(folder,i,j)
            docs = process_directory(j_path,img_store, max_pages=max_pages)
            total_docs.append(docs)
            k+=1
        n+=1
    
    return total_docs

exclude_types = ['Technische FAQ BEG EM','processed_temp']
file_folder = 'data/documents/output/'
file_image_storage = 'data/storage/images'

all_docs = process_all_docs(folder=file_folder,
                 exclude_type=exclude_types,
                 img_store=file_image_storage)


```
