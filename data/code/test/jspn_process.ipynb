```python
list_types = ['Lbl','LBody','L','Li']
elements_lists = [i for i in elements if i['type'] in list_types]
elements_lists
```

```python
from collections import defaultdict
import pandas as pd

# Initialize a dictionary to count the combinations
combinations = defaultdict(int)

for n,i in enumerate(elements):
    type = i['type']
    if n < len(elements)-1:
        type_next = elements[n+1]['type']
        combinations[(type, type_next)] += 1
    else:
        type_next = None
    # print(f'i {type} - i_next {type_next}')

# Convert the dictionary to a DataFrame for better visualization
combination_df = pd.DataFrame(combinations.items(), columns=['Combination', 'Count'])

# Split the combination tuple into separate columns for better readability
combination_df[['Type', 'Next_Type']] = pd.DataFrame(combination_df['Combination'].tolist(), index=combination_df.index)
combination_df = combination_df.drop(columns=['Combination'])

# Rearrange the columns
combination_df = combination_df[['Type', 'Next_Type', 'Count']]
combination_df = combination_df.sort_values(by=['Type', 'Next_Type'])

print(combination_df)
```

```python
lbody_list = []
for n,i in enumerate(elements):
    type = i['type']
    if n < len(elements)-1:
        type_next = elements[n+1]['type']
        i_next = elements[n+1]
        if type_next != 'LBody' and type == 'Lbl':
            dict_temp = {'first':i,'next':i_next}
            lbody_list.append(dict_temp)
    else:
        type_next = None

lbody_list
```

```python
for i in lbody_list:
    first = i['first']['content']
    second = i['next']['content']
    page = i['first']['page']
    type = i['first']['type']
    type_next = i['next']['type']
    print(f'{first} : {second}\npage: {page} - {type}->{type_next}\n')
    # print(f'{i['first']['content']}\n{i['next']['content']}\n')
```

```python
# Headings:

headings = ['Title','H1','H2','H3','H4','H5','H6']

headings_list = []
for n,i in enumerate(elements):
    type = i['type']
    if n < len(elements)-1:
        if type in headings:
            headings_list.append(i)

for i in headings_list:
    first = i['content']
    page = i['page']
    type = i['type']
    print(f'{first}\npage: {page} - {type}\n')
```
